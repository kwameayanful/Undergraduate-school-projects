```python
import numpy as np
import matplotlib.pyplot as plt
import itertools
```

Affine maps of the form $$f(x) = Lx + b$$ where $L$ is a linear map given by a matrix and $b\in\mathbb{R}^2$ where the linear map is a contraction map can be used to produce some very stunning curves. This project uses affine maps to create some graphics.

```python
def deterministic_algorithm(x, b, L, n):
    result, count = [], 0
    num_points = len(x)
    f = lambda x, b, L: np.matmul(L, x) + b
    if len(L) > 1 and len(L) == len(b):
        X, Y = [], []
        X = [x[i][0] for i in range(len(x))]
        Y = [x[i][1] for i in range(len(x))]
        plt.plot(X, Y)
        plt.savefig(f'images/iter00.png')
        plt.show()
        while count < n:
            X, Y = [], []
            start, stop = 0, num_points
            result.append([])
            for j, k in zip(b, L):
                for i in x:
                    result[count].append(f(i, j, k))
            for idx in range(len(result[count])):
                X.append(result[count][idx][0])
                Y.append(result[count][idx][1])
            while stop <= len(X) + 1:
                plt.plot(X[start : stop], Y[start : stop])
                start += num_points
                stop += num_points
            plt.title(f"iteration {count + 1}")
            plt.savefig(f'images/iter{count}.png')
            plt.show()
            x = result[count]
            count += 1
        
    else:
        X, Y = [], []
        X = [x[i][0] for i in range(len(x))]
        Y = [x[i][1] for i in range(len(x))]
        plt.plot(X, Y)
        plt.savefig(f'images/iter00.png')
        plt.show()
        while count < n:
            X, Y = [], []
            start, stop = 0, num_points
            result.append([])
            for j, i  in itertools.product(b, x):
                result[count].append(f(i, j, L))
            for idx in range(len(result[count])):
                X.append(result[count][idx][0])
                Y.append(result[count][idx][1])
            while stop <= len(X) + 1:
                plt.plot(X[start : stop], Y[start : stop])
                start += num_points
                stop += num_points
            plt.title(f"iteration {count + 1}")
            plt.savefig(f'images/iter{count}.png')
            plt.show()
            x = result[count]
            count += 1
```

(a) Use $L_j=\begin{pmatrix}1/3 & 0\\ 0 & 1/3\end{pmatrix}$, $(1\le j\le 5)$ and $b_1=\begin{pmatrix}1/3 \\ 0 \end{pmatrix}$, $b_2=\begin{pmatrix}0 \\ 1/3\end{pmatrix}$, $b_3=\begin{pmatrix}1/3 \\ 1/3 \end{pmatrix}$, $b_4=\begin{pmatrix}2/3 \\ 1/3 \end{pmatrix}$, $b_5=\begin{pmatrix}1/3 \\ 2/3 \end{pmatrix}$ for each iteration $i$, create  the transformation $A_i$ by applying these transformation to $A_{i-1}$. Begin with $A_0$ as the boundary of unit square.

```python
x = [np.array([[0],[0]]), np.array([[1],[0]]), np.array([[1],[1]]), np.array([[0],[1]]), np.array([[0],[0]])]
b = [np.array([[1/3],[0]]), np.array([[0],[1/3]]), np.array([[1/3],[1/3]]), np.array([[2/3],[1/3]]), np.array([[1/3],[2/3]])]
L = np.array([[1/3, 0],[0, 1/3]])

deterministic_algorithm(x, b, L, 6)
```
